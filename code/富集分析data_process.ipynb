```python
def before_GOstat(f1,f2):
    for i in f1.readlines():
        j = i.split('\t')
        for k in j[1].split(','):
            m = j[0] + '\t' + k
            if(m[-1] != '\n'):
                m = m + '\n'
            print(m)
            f2.write(m)
f1 = open('.../swiss_go.out','r')
f2 = open('.../swiss_go_forStats.txt','w')
before_GOstat(f1,f2)
f1.close()
f2.close()
```

```python
# python 解析kegg.json
import json
import re
with open(".../Pipeline/raw_Data/富集分析/ko00001.json","r") as f:    
    fj = f.read()    
    kojson = json.loads(fj)
    with open(".../Pipeline/raw_Data/富集分析/newKegg.tsv", "w") as k:    
        for i in kojson['children']:                
            ii = i['name'].replace(" ", "\t", 1)        
            for j in i['children']:                        
                jj = j['name'].replace(" ", "\t", 1)            
                for m in j['children']:                                
                    if re.findall(r"ko\d{5}",m['name']):                    
                        mm = "ko" + m['name'].replace(" ", "\t", 1)                
                    else :                    
                        mm = m['name'].replace(" ", "\t", 1)                
                    try :                    
                        for n in m['children']:                                                
                            if ";" in n['name']:                            
                                nn = n['name'].replace(" ", "\t", 1).replace("; ", "\t", 1)                        
                            else:                            
                                nn = n['name'].replace(" ", "\t \t", 1)                        
                            k.write(ii + "\t" + jj + "\t" + mm + "\t" + nn + "\n")                
                    except :                                                                                        
                        nn = " \t \t "                    
                        k.write(ii + "\t" + jj + "\t" + mm + "\t" + nn + "\n")
```
